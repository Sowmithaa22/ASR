#**ASR EXPERIMENTS**

###*MODULE 1: IMPLEMENTATION OF ASR MODELS*


```python
#1. HuBERT
import librosa
import torch
from transformers import HubertForCTC, AutoProcessor

# Path of audio file to be transcribed
AUDIO_FILE = r'/content/harvard (1).wav'
# Load the model and tokenizer
processor = AutoProcessor.from_pretrained("facebook/hubert-large-ls960-ft")
model = HubertForCTC.from_pretrained("facebook/hubert-large-ls960-ft")

# Importing the file
speech, rate = librosa.load(AUDIO_FILE, sr=16000)

# Tokenizing the input
inputs = processor(speech, return_tensors="pt", sampling_rate=rate).input_values

# Model logits
with torch.no_grad():
	logits = model(inputs).logits
predicted_ids = torch.argmax(logits, dim=-1)

# Print transcription
transcription = processor.batch_decode(predicted_ids)
print(transcription)
```

```python
#2. Wave2Vec2.0CTC
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
import torch
import librosa

# Load pre-trained model and processor
model_name = "facebook/wav2vec2-large-960h"
processor = Wav2Vec2Processor.from_pretrained(model_name)
model = Wav2Vec2ForCTC.from_pretrained(model_name)

# Function to transcribe audio
def transcribe_audio(file_path):
    # Load audio
    speech, sr = librosa.load(file_path, sr=16000)
    # Process audio
    input_values = processor(speech, return_tensors="pt", sampling_rate=sr).input_values
    # Get model prediction
    with torch.no_grad():
        logits = model(input_values).logits
    # Decode prediction
    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = processor.decode(predicted_ids[0])
    return transcription

# Transcribe an audio file
audio_file = "/content/harvard (1).wav"
transcription = transcribe_audio(audio_file)
print(f'Transcription: {transcription}')
```

```python
import torch
import torchaudio
from transformers import WhisperProcessor, WhisperForConditionalGeneration

# Load the Whisper model and processor
model_name = "openai/whisper-large"
processor = WhisperProcessor.from_pretrained(model_name)
model = WhisperForConditionalGeneration.from_pretrained(model_name)

# Set the model to evaluation mode
model.eval()
import requests

# Function to load and preprocess the audio
def load_audio(url, sample_rate=16000):
    # Download the audio file
    response = requests.get(url)
    audio_path = "temp_audio.mp3"
    with open(audio_path, 'wb') as f:
        f.write(response.content)

    # Load the audio file
    speech_array, sampling_rate = torchaudio.load(audio_path)
    resampler = torchaudio.transforms.Resample(sampling_rate, sample_rate)
    speech = resampler(speech_array).squeeze().numpy()
    return speech

# Example audio file URL
audio_url = "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac"

# Load and preprocess the audio
speech = load_audio(audio_url)

# Preprocess the audio input
inputs = processor(speech, return_tensors="pt", sampling_rate=16000)
# Generate transcription using the model
with torch.no_grad():
    generated_ids = model.generate(inputs["input_features"])

# Decode the generated IDs to text
transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)

# Print the transcription
print("Transcription:", transcription[0])
```

##*MODULE 2: FINE-TUNING ASR MODELS*

```python
import torch
import torchaudio
from datasets import load_dataset
from transformers import WhisperProcessor, WhisperForConditionalGeneration, Seq2SeqTrainingArguments, Seq2SeqTrainer

# Set device to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the Whisper model and processor
model_name = "openai/whisper-large"
processor = WhisperProcessor.from_pretrained(model_name)
model = WhisperForConditionalGeneration.from_pretrained(model_name).to(device)

# Example of loading a dataset from Hugging Face datasets library
dataset = load_dataset("librispeech_asr", "clean", split="train.100")

# Function to preprocess audio files and prepare the dataset
def preprocess_function(examples):
    audio = examples["audio"]

    # Resample audio to 16kHz
    resampler = torchaudio.transforms.Resample(orig_freq=audio["sampling_rate"], new_freq=16000)
    speech = resampler(torch.tensor(audio["array"])).numpy()

    # Prepare inputs and targets
    inputs = processor(speech, sampling_rate=16000, return_tensors="pt").input_features[0]
    with processor.as_target_processor():
        targets = processor(examples["text"]).input_ids
    return {"input_features": inputs, "labels": targets}

# Apply the preprocessing function to the dataset
dataset = dataset.map(preprocess_function, remove_columns=["audio", "text"])
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=3,
    predict_with_generate=True,
    logging_dir='./logs',
)

# Initialize the Seq2SeqTrainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    eval_dataset=dataset,
    tokenizer=processor.feature_extractor,
    data_collator=None,
)
# Start training
trainer.train()

# Evaluate the model on the evaluation set
eval_results = trainer.evaluate()
print("Evaluation Results:", eval_results)
```

```python
import torch
import torchaudio
from datasets import load_dataset, load_metric
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC, Trainer, TrainingArguments

# Set device to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the Wav2Vec2 model and processor
model_name = "facebook/wav2vec2-large-960h"
processor = Wav2Vec2Processor.from_pretrained(model_name)
model = Wav2Vec2ForCTC.from_pretrained(model_name).to(device)
# Example of loading a dataset from Hugging Face datasets library
dataset = load_dataset("librispeech_asr", "clean", split="train.100")

# Function to preprocess audio files and prepare the dataset
def preprocess_function(examples):
    audio = examples["audio"]

    # Resample audio to 16kHz
    resampler = torchaudio.transforms.Resample(orig_freq=audio["sampling_rate"], new_freq=16000)
    speech = resampler(torch.tensor(audio["array"])).squeeze().numpy()

    # Prepare inputs and targets
    inputs = processor(speech, sampling_rate=16000, return_tensors="pt", padding=True)
    with processor.as_target_processor():
        targets = processor(examples["text"]).input_ids
    return {"input_values": inputs.input_values[0], "labels": targets}

# Apply the preprocessing function to the dataset
dataset = dataset.map(preprocess_function, remove_columns=["audio", "text"])
# Load metric for evaluation
wer_metric = load_metric("wer")

def compute_metrics(pred):
    pred_logits = pred.predictions
    pred_ids = torch.argmax(pred_logits, dim=-1)
    pred_str = processor.batch_decode(pred_ids)
    # we do not want to group tokens when computing the metrics
    label_str = processor.batch_decode(pred.label_ids, group_tokens=False)
    wer = wer_metric.compute(predictions=pred_str, references=label_str)
    return {"wer": wer}

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=3,
    logging_dir='./logs',
    logging_steps=10,
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    eval_dataset=dataset,
    tokenizer=processor.feature_extractor,
    data_collator=processor.default_data_collator,
    compute_metrics=compute_metrics
)
# Start training
trainer.train()
# Evaluate the model on the evaluation set
eval_results = trainer.evaluate()
print("Evaluation Results:", eval_results)
```
